```python
import pandas as pd 

df = pd.read_excel("MOB-EQ-00090.xlsx",  converters={'Valor de Custo Unitario': float, "CUSTO ORIGEM":float, "DEP ORIGEM":float, "RES ORIGEM":float, })
df
```

```python
dataitem = df.iloc[:, 0:8]
dataitem
```

```python
dataOrigem = df.iloc[0:1,10:]
dataOrigem
```

```python
newdf = pd.DataFrame(columns=dataitem.columns)

for i, iter in dataitem.iterrows():
    quantity = int(iter["quantidade"])
    for i in range(quantity):
        newdf.loc[len(newdf)] = iter
newdf
```

```python
newdf["Valor de Custo Unitario"].sum()
```

```python

porcents = []
soma = newdf["Valor de Custo Unitario"].sum()

for i, iter in newdf.iterrows():
    valor = iter["Valor de Custo Unitario"] / soma
    valor = float(valor)
    porcents.append(valor)

newdf["porcents"] = porcents
newdf
```

```python
CustoNotaTranform = []
depreciationNotaTranform = []
residualNotaTranform = []


for i, iter in newdf.iterrows():
   valuecoast = dataOrigem["CUSTO ORIGEM"][0] * iter["porcents"] 
   valueDep = dataOrigem["DEP ORIGEM"][0] * iter["porcents"]

   CustoNotaTranform.append(round(valuecoast, 2))
   depreciationNotaTranform.append(round(valueDep, 2))
   residualNotaTranform.append(round(valuecoast - valueDep, 2))

newdf["CUSTO NOTA"] = CustoNotaTranform
newdf["DEPRECIAÇÃO NOTA"] = depreciationNotaTranform
newdf["RESIDUAL NOTA"] = residualNotaTranform
```

```python
difCusto = float(newdf["CUSTO NOTA"].sum() - dataOrigem["CUSTO ORIGEM"])
difdeprec = float(newdf["DEPRECIAÇÃO NOTA"].sum() - dataOrigem["DEP ORIGEM"])
difresidual = float(newdf["RESIDUAL NOTA"].sum() - dataOrigem["RES ORIGEM"])

newdf["DEPRECIAÇÃO NOTA"].sum()
index = len(newdf) - 1

newdf["CUSTO NOTA"][index] = newdf["CUSTO NOTA"][index] - difCusto
newdf["DEPRECIAÇÃO NOTA"][index] = newdf["DEPRECIAÇÃO NOTA"][index] - difdeprec
newdf["RESIDUAL NOTA"][index] = newdf["RESIDUAL NOTA"][index] - difresidual



newdf
```

```python
note = str(newdf["Bem origem "][0])

newdf.to_excel(f"subir/subir-{note}2.xlsx", index=False)
```
